```python
tmp_phns = np.concatenate([phns_layer_0, phns_layer_1, phns_layer_2])
tmp_spkrs = np.concatenate([spkrs_layer_0, spkrs_layer_1, spkrs_layer_2])
```

```python
def b(a, mi=None, ma=None):
    if mi != None and ma != None:
        return (a - mi) / (ma - mi)
    else:
        return (a - a.min()) / (a.max() - a.min())

phns = b(phns_layer_0, tmp_phns.min(), tmp_phns.max()) #* 255
spkrs = b(spkrs_layer_0, tmp_spkrs.min(), tmp_spkrs.max()) #* 255
img = [(phns[i],0,spkrs[i]) for i in xrange(phns.shape[0])]
img = np.array(img)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
labels = ["layer 1","layer 2","layer 3","word emb","spkr emb"]
ax.set_xticks([0.4 + i for i in xrange(len(labels))])
ax.set_xticklabels(labels)
ax.set_yticklabels([])

phns_med = np.median(tmp_phns)
spkrs_med = np.median(tmp_spkrs)

for i, (phn_layer, spkr_layer) in enumerate([(phns_layer_0, spkrs_layer_0), (phns_layer_1, spkrs_layer_1), (phns_layer_2, spkrs_layer_2)]):
    phns = phn_layer > phns_med
    spkrs = spkr_layer > spkrs_med
    none = - (phns + spkrs)
    both = phns * spkrs
    phns2 = phns * (- spkrs)
    spkrs2 = spkrs * (- phns)
    print np.sum(none), np.sum(phns2), np.sum(spkrs2), np.sum(both)
    nones = np.sum(none)
    phnss = np.sum(phns2) + nones
    boths = np.sum(both) + phnss
    spkrss = np.sum(spkrs2) + boths
    print nones, phnss, spkrss, boths
    bar([i],[spkrss], color='b')
    bar([i],[boths], color='purple')
    bar([i],[phnss], color='r')
    bar([i],[nones], color='black')

#phns_emb_med = np.median(np.concatenate([phns_emb_word, spkrs_emb_word]))
#spkrs_emb_med = np.median(np.concatenate([phns_emb_spkr, spkrs_emb_spkr]))
phns_emb_med = np.median(np.concatenate([phns_emb_word, phns_emb_spkr]))
spkrs_emb_med = np.median(np.concatenate([spkrs_emb_word, spkrs_emb_spkr]))

for i, (phn_layer, spkr_layer) in enumerate([(phns_emb_word, spkrs_emb_word), (phns_emb_spkr, spkrs_emb_spkr)]):
    phns = phn_layer > phns_emb_med
    spkrs = spkr_layer > spkrs_emb_med
    none = - (phns + spkrs)
    both = phns * spkrs
    phns2 = phns * (- spkrs)
    spkrs2 = spkrs * (- phns)
    print np.sum(none), np.sum(phns2), np.sum(spkrs2), np.sum(both)
    nones = np.sum(none) * 5
    phnss = np.sum(phns2) * 5 + nones
    boths = np.sum(both) * 5 + phnss
    spkrss = np.sum(spkrs2) * 5 + boths
    print nones, phnss, spkrss, boths
    bar([i+3],[spkrss], color='b')
    bar([i+3],[boths], color='purple')
    bar([i+3],[phnss], color='r')
    bar([i+3],[nones], color='black')
```
